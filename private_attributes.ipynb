Private attributes can be accsessible within the class, but not outside of the class. They are denoted by a double underscore prefix(like __private_attribute).

```python
class Account:
    def __init__(self, ac_name, ac_password):
        self.ac_name = ac_name
        self.__ac_password = ac_password
    
    def show_password(self):
        print(f"Your password is: {self.__ac_password}")


account1 = Account("John", "12345")
print(account1.ac_name)  # Output: John
account1.show_password()  # Output: Your password is: 12345

print(account1.__ac_password)  # This will raise an AttributeError because __ac_password is a private attribute
```
