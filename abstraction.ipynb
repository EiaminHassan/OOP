Abstraction means hiding internal details and showing only the necessary features of an object.

To create a abstract class in Python, we can use the abc module. An abstract class is a class that cannot be instantiated and is meant to be subclassed. It can contain abstract methods, which are methods that are declared but not implemented in the abstract class. Subclasses of the abstract class must implement the abstract methods.

```python
# create an abstract class.
from abc import ABC, abstractmethod

class BankApp(ABC):

    def database_connection(self):
        print("Connecting to the database...")
    
    @abstractmethod
    def login(self):
        pass
```

```python
# create a subclass of the abstract class and implement the abstract method.
class MyBankApp(BankApp):

    def login(self):
        print("Logging in to the bank app...")

obj = MyBankApp()
obj.database_connection()
obj.login()
```
