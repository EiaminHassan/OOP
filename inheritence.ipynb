- Inheritance allows us to define a class that inherits all the methods and properties from another class.

- **Parent class** is the class being inherited from, also called base class.

- **Child class** is the class that inherits from another class, also called derived class.

```python
# Parent Class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

# x = Person("Alice", 30)
# x.display_info()


# Child Class
class Student(Person):
    pass # This means that the Student class inherits everything from the Person class without adding anything new.

x = Student("Bob", 20)
x.display_info()  # This will call the display_info method from the Person class, since Student inherits from Person.
```

Add \__init__ method to the child class

```python
# Parent Class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")


# Child Class
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id
    def display_info(self):
        super().display_info()  # Call the display_info method from the Person class
        print(f"Student ID: {self.student_id}")

x = Student("Bob", 20, "S12345")
x.display_info()  # This will call the display_info method from the Person class, since Student inherits from Person.
```
