**Methods are functions that belong to a class**

```python
class Person:
    def __init__(self,name):
        self.name = name
    
    def say_hello(self):
        print(f"Hello, my name is {self.name}")

p1 = Person("Alice")
p1.say_hello()
```

**Methods with parameters:** Methods can accept parameters like regular functions.

```python
class Calculator:
    def add(self,a ,b):
        return a+b

calc = Calculator()
print(calc.add(5, 3))
```

**Methods accesing properties:** A method that accesses object properties.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def get_info(self):
        return f"{self.name} is {self.age} years old."

p = Person("Bob", 30)
print(p.get_info())
```

**Methods Modifying Properties:** A method that modifies object properties.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def modify_age(self):
        self.age += 1
        print(f"{self.name} is now {self.age} years old.")

p = Person("Charlie", 25)
p.modify_age()
p.modify_age()
```

**The \__str()__ method:** A special method that controls what is returned when the object is printed.

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("Emil", 36)
print(p1)
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def __str__(self):
    return f"{self.name} ({self.age})"

p1 = Person("Tobias", 36)
print(p1)
```

**Example of a class having multiple methods**

```python
class Players:
    def __init__(self, name):
        self.team_name = name
        self.my_players = []

    def add_players(self,player):
        self.my_players.append(player)
    
    def list_players(self):
        print(f"Players in {self.team_name}:")
        for player in self.my_players:
            print(player)

team = Players("My Team")
team.add_players("Messi")
team.add_players("Neymar")
team.add_players("Mbappe")
team.add_players("Pedri")
team.add_players("Gavi")
team.list_players()
```
